## Copy the data for safe use

Create a new column named `'ref'` for the **id** i will use in the **foreign key** in the fact table

```python
dfc = dfc.reset_index(drop=False)  

dfc.rename(columns={'index': 'ref'}, inplace=True)
```

```python
dfc.head(5)
```

### Now our data is cleaned lets **transform** it in the shape of our Dimensions and fact tables :

![Model](./Images/Model.png)


## Date Dimension

```python
# Create dim_date
dfc['Date_ID']  = dfc['ref'] 
dfc['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')  # make sure of that the Date in datetime format
dfc['Day'] = dfc['Date'].dt.day
dfc['Day_Name'] = dfc['Date'].dt.day_name()
dfc['Month'] = dfc['Date'].dt.month
dfc['Month_Name'] = dfc['Date'].dt.month_name()
dfc['Quarter'] = dfc['Date'].dt.quarter
dfc['Year'] = dfc['Date'].dt.year

dim_date = dfc[['Date_ID','Date', 'Day', 'Day_Name', 'Month', 'Month_Name', 'Quarter', 'Year']].reset_index(drop=True)
```

```python
dim_date
```

## Time Dimension

```python
# Create dim_time
# Convert 'Time' column to string
dfc['Time'] = dfc['Time'].astype(str)

# Use regex to extract hour and minute
dfc[['Hour', 'Minute']] = dfc['Time'].str.extract(r'(\d{2}):(\d{2}):\d{2}')

# Convert extracted values to integer
dfc['Hour'] = dfc['Hour'].astype(int)
dfc['Minute'] = dfc['Minute'].astype(int)
```

Here i just transfer the `time` to string because i faced an error using the datetime format extracting `hour`, `minute` standard way

```python
dfc['Time_ID']  = dfc['ref']
```

```python
dim_time = dfc[['Time_ID','Time', 'Hour', 'Minute']].reset_index(drop=True)
```

```python
dim_time
```

## Location Dimension

```python
# Create dim_location
dfc['Location_ID']  = dfc['ref']
dim_location = dfc[['Location_ID','Easting', 'Northing', 'Location', 'Local_Authority']].reset_index(drop=True)
```

```python
dim_location
```

## Severity Dimension

```python
# Create dim_severity
dfc['Severity_ID']  = dfc['ref']
dim_severity = dfc[['Severity_ID','Severity']].reset_index(drop=True)
```

```python
dim_severity
```

## Road Condition Dimension

```python
# Create dim_road_condition
dfc['Road_Condition_ID']  = dfc['ref']
dim_road_condition = dfc[['Road_Condition_ID','Road_Condition', 'Weather', 'Visibility', 'Speed_Limit']].reset_index(drop=True)
```

```python
dim_road_condition
```

## Crossing Dimension

```python
# Create dim_crossing
dfc['Crossing_ID']  = dfc['ref']
dim_crossing = dfc[['Crossing_ID','Crossing_Facility', 'Crossing_Control_Type']].reset_index(drop=True)
```
